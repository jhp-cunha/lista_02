**Questão 01 - Usando o mesmo portfolio da sua lista anterior, normalize os retornos logarítmicos,( z-score). O que ele significa?**

```python
import pandas as pd
df = pd.read_excel('F:/Quant/Listas/base.lista1.xlsx',header=3,skiprows=0,index_col=0)
df = df.apply(pd.to_numeric, errors='coerce')
df.index = pd.to_datetime(df.index, errors = 'coerce')
df = df.dropna()
df.index.name = 'Data'
df.index
df.columns
tickers = df.columns.tolist()

df2 = df.copy()
df2 =df2.dropna()
valor1 = df2['BBDC4'].dropna().iloc[0]
df2['BBDC4'] = pd.to_numeric(df['BBDC4'], errors='coerce')
df2['index_BBDC4'] = df2['BBDC4']/valor1
df2['index_BBDC4']

valor2 = df2['BBAS3'].dropna().iloc[0]
df2['BBAS3'] = pd.to_numeric(df['BBAS3'], errors='coerce')
df2['index_BBAS3'] = df2['BBAS3']/valor2
df2['index_BBAS3']

valor3 = df2['BPAC5'].dropna().iloc[0]
df2['BPAC5'] = pd.to_numeric(df['BPAC5'], errors='coerce')
df2['index_BPAC5'] = df2['BPAC5']/valor3
df2['index_BPAC5']

valor4 = df2['ITUB4'].dropna().iloc[0]
df2['ITUB4'] = pd.to_numeric(df['ITUB4'], errors='coerce')
df2['index_ITUB4'] = df2['ITUB4']/valor4
df2['index_ITUB4']

valor5 = df2['SANB4'].dropna().iloc[0]
df2['SANB4'] = pd.to_numeric(df['SANB4'], errors='coerce')
df2['index_SANB4'] = df2['SANB4']/valor5
df2['index_SANB4']

valor6 = df2['IBOV'].dropna().iloc[0]
df2['IBOV'] = pd.to_numeric(df['IBOV'], errors='coerce')
df2['index_IBOV'] = df2['IBOV']/valor6
df2['index_IBOV']

import numpy as np

df2['log return BBDC4'] = np.log(df2['index_BBDC4']) - np.log(df2['index_BBDC4'].shift(1))
df2['log return BBAS3'] = np.log(df2['index_BBAS3']) - np.log(df2['index_BBAS3'].shift(1))
df2['log return BPAC5'] = np.log(df2['index_BPAC5']) - np.log(df2['index_BPAC5'].shift(1))
df2['log return ITUB4'] = np.log(df2['index_ITUB4']) - np.log(df2['index_ITUB4'].shift(1))
df2['log return SANB4'] = np.log(df2['index_SANB4']) - np.log(df2['index_SANB4'].shift(1))
df2['log return IBOV'] = np.log(df2['index_IBOV']) - np.log(df2['index_IBOV'].shift(1))
```

```python
### Escreva aqui seu código para responder a questão 1.



df2['mean log return BBDC4'] = df2['log return BBDC4'].mean()
df2['mean log return BBAS3'] = df2['log return BBAS3'].mean()
df2['mean log return BPAC5'] = df2['log return BPAC5'].mean()
df2['mean log return ITUB4'] = df2['log return ITUB4'].mean()
df2['mean log return SANB4'] = df2['log return SANB4'].mean()
df2['mean log return IBOV'] = df2['log return IBOV'].mean()

df2['std log return BBDC4'] = df2['log return BBDC4'].std()
df2['std log return BBAS3'] = df2['log return BBAS3'].std()
df2['std log return BPAC5'] = df2['log return BPAC5'].std()
df2['std log return ITUB4'] = df2['log return ITUB4'].std()
df2['std log return SANB4'] = df2['log return SANB4'].std()
df2['std log return IBOV'] = df2['log return IBOV'].std()

df2['z-score log return BBDC4'] = (df2['log return BBDC4'] - df2['mean log return BBDC4']) / df2['std log return BBDC4']
df2['z-score log return BBAS3'] = (df2['log return BBAS3'] - df2['mean log return BBAS3']) / df2['std log return BBAS3']
df2['z-score log return BPAC5'] = (df2['log return BPAC5'] - df2['mean log return BPAC5']) / df2['std log return BPAC5']
df2['z-score log return ITUB4'] = (df2['log return ITUB4'] - df2['mean log return ITUB4']) / df2['std log return ITUB4']
df2['z-score log return SANB4'] = (df2['log return SANB4'] - df2['mean log return SANB4']) / df2['std log return SANB4']
df2['z-score log return IBOV'] = (df2['log return IBOV'] - df2['mean log return IBOV']) / df2['std log return IBOV']

print(df2[['z-score log return BBDC4', 'z-score log return BBAS3', 'z-score log return BPAC5', 'z-score log return ITUB4', 'z-score log return SANB4', 'z-score log return IBOV']])
```

O Z-Score fornece a distância medida em desvios-padrão da amostra entre uma varíavel e a média amostral, assim conseguindo padronizar os dados da amostra

**Questão 02 - Com os retornos normalizados, plote um histograma deles. Também adicione uma curva de gauss e 3 curvas de t-student com n = 5,10 e 50. O que difere a t-student da curva gaussiana?**

```python
### Escreva aqui seu código para responder a questão 2.


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import norm, std
from scipy.stats import norm,t

#código para histogramas sem curvas de gauss etc.

plt.hist(df2['log return BBDC4'], bins=20, label='BBDC4', alpha=0.5, density=True)

#calculos para curva de gauss

df2['mean log return BBDC4'] = df2['log return BBDC4'].mean()


df2['std log return BBDC4'] = df2['log return BBDC4'].std()


x = np.linspace(df2['log return BBDC4'].min(), df2['log return BBDC4'].max(), 100)


plt.plot(x, norm.pdf(x, df2['mean log return BBDC4'].values[0], df2['std log return BBDC4'].values[0]), 'r-', linewidth=2, label='Gaussian')

#fazendo a curva t-student para n=5, n=10 e n=50

plt.plot(x, t.pdf(x, df=5, loc=df2['mean log return BBDC4'].iloc[0], scale=df2['std log return BBDC4'].iloc[0]), 'g--', linewidth=2, label='t-student (n=5)')
plt.plot(x, t.pdf(x, df=10, loc=df2['mean log return BBDC4'].iloc[0], scale=df2['std log return BBDC4'].iloc[0]), 'b--', linewidth=2, label='t-student (n=10)')
plt.plot(x, t.pdf(x, df=50, loc=df2['mean log return BBDC4'].iloc[0], scale=df2['std log return BBDC4'].iloc[0]), 'm--', linewidth=2, label='t-student (n=50)')


plt.title('BBDC4 -Log Returns')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.legend(loc='upper right')
plt.legend()
plt.show()
```

**Questão 04 - Plote a PDF e a CDF dos log-retornos do seu portfólio.**

```python
### Escreva aqui seu código para responder a questão 4.

CORES = {
    "BBDC4": "#5b7fa6",
    "BBAS3": "#2ecc71",
    "BPAC5": "#e74c3c",
    "ITUB4": "#f39c12",
    "SANB4": "#9b59b6",
    "IBOV": "#ff00ff"
}

fig, axes = plt.subplots(6, 2, figsize=(14, 22))


for i, ticker in enumerate(tickers):
    r    = df2[f'log return {ticker}']
    med  = r.mean()
    std  = r.std()
    cor  = CORES[ticker]
    nome = ticker
    x    = np.linspace(r.min() - 0.01, r.max() + 0.01, 500)

    # PDF -> é a função de densidade de probabilidade, ou seja, aponta a probabilidade relativa de uma variável (nesse caso log retornos de ações e do IBOV) assumir um valor específico
    ax_pdf = axes[i, 0]
    ax_pdf.hist(r, bins=60, density=True, alpha=0.5,
                color=cor, edgecolor="white", label="Dados reais")
    ax_pdf.plot(x, norm.pdf(x, med, std), color="#2c3e50", lw=2, label="Normal teórica")
    ax_pdf.set_title(f"{nome} — PDF", fontsize=11)
    ax_pdf.set_xlabel("Retorno log.")
    ax_pdf.set_ylabel("Densidade")
    ax_pdf.legend(fontsize=9)

    # CDF -> é a função de distribuição acumulada, ou seja, aponta a probabilidade de uma variável assumi valor menor ou igual a um valor específico
    ax_cdf = axes[i, 1]
    r_sorted = np.sort(r)
    cdf_emp  = np.arange(1, len(r_sorted) + 1) / len(r_sorted)
    ax_cdf.plot(r_sorted, cdf_emp, color=cor, lw=2, label="CDF empírica")
    ax_cdf.plot(x, norm.cdf(x, med, std), color="#2c3e50", lw=2,
                linestyle="--", label="CDF Normal teórica")
    ax_cdf.set_title(f"{nome} — CDF", fontsize=11)
    ax_cdf.set_xlabel("Retorno log.")
    ax_cdf.set_ylabel("Prob. acumulada")
    ax_cdf.legend(fontsize=9)

plt.tight_layout()
plt.savefig("exercicio2_pdf_cdf.png", dpi=130, bbox_inches="tight")
plt.show()

### Para responder a parte interpretativa, crie uma nova célula de texto!
```
