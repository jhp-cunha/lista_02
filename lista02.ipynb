**Questão 01 - Usando o mesmo portfolio da sua lista anterior, normalize os retornos logarítmicos,( z-score). O que ele significa?**

```python
import pandas as pd
import numpy as np

df = pd.read_excel('F:/Quant/Listas/base.lista1.xlsx',header=3,skiprows=0,index_col=0)
df = df.apply(pd.to_numeric, errors='coerce')
df.index = pd.to_datetime(df.index, errors = 'coerce')
df = df.dropna()
df.index.name = 'Data'
df.index
df.columns
tickers = df.columns.tolist()

df2 = df.copy()
df2 =df2.dropna()

for ticker in tickers:
    index = df2[ticker].dropna().iloc[0]
    df2[ticker] = pd.to_numeric(df[ticker], errors='coerce')
    df2['index_' + ticker] = df2[ticker]/index
    valor1 = df2['BBDC4'].dropna().iloc[0]
    df2['log return' + ' ' + ticker] = np.log(df2['index_' + ticker]) - np.log(df2['index_' + ticker].shift(1))
    print(df2['log return' + ' ' + ticker])



```

```python
### Escreva aqui seu código para responder a questão 1.


for ticker in tickers:
    df2['z-score log return' + ' ' + ticker] = (df2['log return' + ' ' + ticker] - df2['log return' + ' ' + ticker].mean()) / df2['log return' + ' ' + ticker].std()
    z_score_value = df2['z-score log return' + ' ' + ticker].iloc[-1]
    print(f"Z-score log return {ticker}: {z_score_value:.3f}")
```

O Z-Score fornece a distância medida em desvios-padrão da amostra entre uma varíavel e a média amostral, assim conseguindo padronizar os dados da amostra

**Questão 04 - Plote a PDF e a CDF dos log-retornos do seu portfólio.**

```python
### Escreva aqui seu código para responder a questão 4.

CORES = {
    "BBDC4": "#5b7fa6",
    "BBAS3": "#2ecc71",
    "BPAC5": "#e74c3c",
    "ITUB4": "#f39c12",
    "SANB4": "#9b59b6",
    "IBOV": "#ff00ff"
}

fig, axes = plt.subplots(6, 2, figsize=(14, 22))


for i, ticker in enumerate(tickers):
    r    = df2[f'log return {ticker}']
    med  = r.mean()
    std  = r.std()
    cor  = CORES[ticker]
    nome = ticker
    x    = np.linspace(r.min() - 0.01, r.max() + 0.01, 500)

    # PDF -> é a função de densidade de probabilidade, ou seja, aponta a probabilidade relativa de uma variável (nesse caso log retornos de ações e do IBOV) assumir um valor específico
    ax_pdf = axes[i, 0]
    ax_pdf.hist(r, bins=60, density=True, alpha=0.5,
                color=cor, edgecolor="white", label="Dados reais")
    ax_pdf.plot(x, norm.pdf(x, med, std), color="#2c3e50", lw=2, label="Normal teórica")
    ax_pdf.set_title(f"{nome} — PDF", fontsize=11)
    ax_pdf.set_xlabel("Retorno log.")
    ax_pdf.set_ylabel("Densidade")
    ax_pdf.legend(fontsize=9)

    # CDF -> é a função de distribuição acumulada, ou seja, aponta a probabilidade de uma variável assumi valor menor ou igual a um valor específico
    ax_cdf = axes[i, 1]
    r_sorted = np.sort(r)
    cdf_emp  = np.arange(1, len(r_sorted) + 1) / len(r_sorted)
    ax_cdf.plot(r_sorted, cdf_emp, color=cor, lw=2, label="CDF empírica")
    ax_cdf.plot(x, norm.cdf(x, med, std), color="#2c3e50", lw=2,
                linestyle="--", label="CDF Normal teórica")
    ax_cdf.set_title(f"{nome} — CDF", fontsize=11)
    ax_cdf.set_xlabel("Retorno log.")
    ax_cdf.set_ylabel("Prob. acumulada")
    ax_cdf.legend(fontsize=9)

plt.tight_layout()
plt.savefig("exercicio2_pdf_cdf.png", dpi=130, bbox_inches="tight")
plt.show()

### Para responder a parte interpretativa, crie uma nova célula de texto!
```

**Questão 05 - Encontre kurtosis e skewness do retornos logarítmicos do seu portfólio. Qual o significado dessas medidas? Quais são os valores para uma normal padrão e qual o significado da diferença desses valores para o seu portfólio e para uma normal padrão?**


```python
### Escreva aqui seu código para responder a questão 5.

from numpy import var
from scipy import stats
from scipy.stats import norm, skew, kurtosis

print(f"\n{'Ação':<14} {'Skewness':>10} {'Kurtosis':>10}")
print("-" * 45)

resultados_ex1 = {}
for ticker in tickers:
    r = df2['log return ' + ticker].dropna()
    skw  = skew(r)
    kurt = kurtosis(r)
    sk = dict(skewness=skw, kurtosis=kurt)          
    print(f"{ticker:<14} {skw:>10.4f} {kurt:>10.4f}")

### Para responder a parte interpretativa, crie uma nova célula de texto!
```

A skewness (assimetria) mede justamente a assimetria da distribuição de valores de uma variável ao redor da média, ou seja, o quanto o gráfico foge de uma distribuição de Gauss. Caso Skewness > 0, isso significa que há mais observações na esquerda do gráfico, ou seja abaixo da média e vice-versa. No caso de Skewness = 0, o gráfico possui distribuição normal. No exemplo do portfólio, todos os gráficos de log return dos ativos possuem skewness negativa, assim há mais retornos acima da média.
A kurtosis (curtose) avalia a propensão de outliers em uma amostra, ou seja, quão achatadas são as caudas do gráfico de distribuição. Em caso de Kurtosis = 0 (curva mesocúrtica), há o cenário de distribuição normal, ou seja, muitos valores próximos à média e poucos outliers. Caso Kurtosis > 0 (curva leptocúrtica), há bastante valores próximos à média, no entanto também há forte presença de outliers. Por último, Kurtosis < 0, há poucos dados próximos à média e poucos outliers, sendo assim uma amostra extremamente distribuída. Dentro do contexto do portfólio, todos os ativos apresentaram kurtosis positiva, logo há sim valores próximos à média, mas também há outliers, sendo assim um cenário condizente com o mundo de produtos de renda variável.

**Questão 08 - Teste se os retornos do portfólio são estatisticamente diferentes de zero e do retorno do índice BOVESPA. Qual alpha você utilizaria?**

```python
print("\nH₀: µ = 0 |  H₁: µ ≠ 0 | Bilateral | α = 5%")
print(f"\n{' ':<16} {'Média':>10} {'Erro Pad.':>10} {'Z':>8} {'p-valor':>10} {'Decisão'}")
print("-" * 72)

alpha     = 0.05
z_critico = norm.ppf(1 - alpha/2)

r = df2['log return a.a. portfolio'].dropna()
n    = len(r)
med  = r.mean()
std  = r.std()
ep   = std / np.sqrt(n)
z    = med / ep
pval = 2 * (1 - norm.cdf(abs(z)))
dec  = "REJEITA H₀" if pval <= alpha else "Não rejeita H₀"
print(f"{'Retorno Ativo':<16} {med:>10.6f} {ep:>10.6f} {z:>8.3f} {pval:>10.4f}  {dec}")

print(f"\n(Valor crítico: ±{z_critico:.4f})\n")
```
